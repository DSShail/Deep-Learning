```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
sns.scatterplot(x=df['cgpa'],y=df['resume_score'],hue=df['placed'])
```

```python
X=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.linear_model import Perceptron
p=Perceptron()
```

#### z=W1X1+W2X2+B
- W1 is cgpa
- W2 is resume score
- b is bias


**We are going to calculate W1 and W2 using perceptron**

```python
p.fit(X,y)
```

#### W1 and W2 are weights

```python
p.coef_
```

#### b-BIAS

```python
p.intercept_
```
